### 1. **Simple Calculator**

* Take two numbers as input (handle type casting).
* Ask the user to choose an operation (`+`, `-`, `*`, `/`, `%`).
* Perform the operation using conditional statements and print the result.
* Use loops to allow multiple calculations until the user wants to exit.

### 2. **Number Analysis Program**

* Take an integer input.
* Use loops and conditional statements to:

  * Check if the number is prime.
  * Calculate factorial.
  * Print the sum of digits.
  * Check if palindrome.
* Use functions for each sub-task.

---

### 3. **String Manipulation Tool**

* Take a string input from the user.
* Use loops and conditional statements to:

  * Count vowels and consonants.
  * Reverse the string.
  * Check if palindrome.
  * Remove all spaces and replace them with underscores.
  * Convert string to uppercase and lowercase.
* Use string methods and slicing.

### 4. **Student Grade Calculator**

* Take student name, roll number, and marks of 5 subjects as input (use type casting).
* Calculate total and percentage.
* Use conditional statements to assign grades (`A`, `B`, `C`, `D`, `F`).
* Print a formatted report using f-strings.

### 5. **Number Guessing Game**

* Generate a random number between 1 to 100.
* Use loops to allow user multiple guesses.
* Use conditional statements to tell if guess is too high, too low, or correct.
* Use loop control statements (`break` when correct).
* Count the number of attempts.

### 6. **Shopping Bill Calculator**

* Input multiple items with their price and quantity (use lists and loops).
* Calculate total amount.
* Apply discount based on total bill using conditional statements.
* Print a detailed bill with item-wise cost and total amount.

* Take a password string input.

  * Minimum 8 characters.

  * At least one lowercase letter.
  * At least one digit.
  * At least one special character.
* Print appropriate validation messages.
